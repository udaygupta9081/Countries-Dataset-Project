```python
#how many total regions are there
```

```python
re=df1['region'].unique()
count=0
for i in re:
    print(i) 
    count+=1 
print("Total number of different regions are: ",count)
```

```python
df1[df1['political_leader'].isna()]
```

```python
#how many country have Republic in their full name
```

```python
count=0
def republicInName(txt): 
    global count
    if "republic" in txt.lower():
        count+=1
    return txt
df1['country_long'].apply(republicInName)
print(count)
```
